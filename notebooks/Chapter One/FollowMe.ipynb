# Chapter 1: Follow Me – Complete Edits

He walked into Brian's life with very little warning or discretion at all, and though it was not a complete surprise—for Brian had knowledge of his own—a sliver of fear crept its way into his heart. The man had used his *real name* for one, and Brian had never met the man before. It was peculiar indeed. Strangers didn't often know his *real* name. Brian was in for a wild ride, and despite all that would soon come crashing into his life, fate could not have picked a more peculiarly *mundane* setting. Our story begins in a coffee shop one morning, as Brian sat there, checking his email.  

"Brian McCord, what are you doing with your life?" the stranger asked abruptly, mocking him it seemed. No introduction. No opening words. Just a blatant and somewhat rude, attention-grabbing jeer. To Brian the man seemed fully intent on catching him off guard. However, Brian wasn't prone to such things.  

He turned around, half expecting this. He'd felt it coming for some time, though he had not known just when or where such a thing would manifest. When Brian looked up, the man was glaring back at him, eyes green and menacing. As it was not in Brian's nature to wear his emotions on his sleeve, he calmly lowered his sunglasses, meeting the gaze. Brian's steely grey eyes pierced into the man with a matched intensity. He smirked, attempting to throw *the stranger* off of *his* game.  

The stranger frowned knowingly, shaking his head. "Oh, don't give me that! You're absolutely terrified!" he exclaimed, a sinister smile creeping across his face, his voice ending in a hiss. "Do you know *who* I am?" Eyes flaring with some dark and hidden hunger. The stranger moved like a snake, repositioning himself in front of Brian's table.  

At this Brian casually took his sunglasses the rest of the way off, laying them down and taking a drink of his coffee as if bored, but when he looked again, the man was nearly breathing down his throat.  

"It is time!" he hissed, green eyes flashing with lightning it seemed.  

Brian didn't so much as flinch, though... he probably should have. After all, he *hadn't* seen the man move, not even in his peripheral. One moment he'd been a safe distance away... and the next, he'd invaded Brian's personal space—again, intentionally.  

Brian maintained his bored demeanor. "And you sir are?" he inquired nonchalantly.  

An evil and knowing smile formed upon the stranger's lips, raising his eyebrows suggestively as he said, "Oh... Don't you worry about that... Just follow me!" And with that the strange man turned swiftly around, one arm behind his back as he strode confidently towards the door. Upon exiting the cafe, he flung the door open with a shove so forceful it nearly knocked over the elderly couple seeking to enter at that same time. He muttered something Brian couldn't make out, however, that appeared to offend the couple so much that nearly being hit with the door seemed trivial at best. Odder still, he didn't so much as look back to make sure Brian followed, even after such a rude display of theatrics.  

...But Brian did follow. He didn't know why. He didn't even ponder it much. His feet began moving before he realized it, and for some reason, he let them go. It was as if he were being pulled by a current, against which there *could be no fighting*. That is not to say he necessarily *wanted* to combat it. Brian had known of this day for many years, and though he dreaded to see its arrival—partially for reasons he didn't yet understand—something about the man intrigued him. Plus, he was due for a date with a cigarette, and smoking wasn't allowed inside.  

This was the excuse he used to quell the arrogance in his mind, which seemed to disagree with him showing any interest in the least. You see, the *strangeness* of this event was not what led him to gather his laptop, coffee, and half-eaten breakfast. It was not the man's strange movements or speed which caused him to place his belongings into his car, lock it, then look around for the direction he had taken. No, not even the odd glimmer in the man's eyes, which seemed more compelling than his knowledge of Brian's name, had set his feet to walking, lighting a cigarette once he'd spotted him. None of this interested Brian at all. However, it was the *inexplicable* strangeness of the walk itself that caused him to increase his pace, and this was very out of character for the young man.  

Brian had noticed, by chance as it were, just how fast the man was walking and yet—somehow—the distance between them stayed fixed. Brian, you see, began following at his usual slow and steady, almost half-bored pace, unconcerned—or so he thought—with whether or not the man escaped his field of vision. But when Brian made *this* odd observation, he decided it best to test the crazed idea that surfaced in his mind. With all that had just transpired in the last few moments, what could it hurt to just *go with the flow*?  

He began at first to walk slightly faster, gauging his speed with the steps of the strange man, who, as Brian also noted, had not bothered to look back, *not even once* to check on his progress. When it became apparent to him that the man had not sped up, nor that the distance between them had changed in the slightest—for Brian was exactly a full block behind him—he broke out into a slight jog, carefully watching to ensure the man was indeed *still just walking*. Sure enough, the man's gait had not adjusted... but then again, neither had the distance between them.  

This was not the first time something of this nature had appeared in Brian's life. He was twenty-three now, and until this point he'd always experienced periods of time—which fizzled out eventually—wherein it seemed as if the whole of the universe—God, as some may call it—were attempting to tell him... *something*. Several times he'd followed hopelessly after various "signs and wonders," only to be let down in the end, as his extra-sensory perception of the world slowly slipped away into oblivion, leaving him questioning everything all the more.  

He'd written it off as a dream the first time he'd been approached in a like manner. He was only nine at the time, though the man he'd encountered then had been quite the opposite of this *new stranger*. He had made a point of ignoring all of the other odd people who'd come afterward, each saying strange things and beckoning him to follow them. Despite this he'd never managed to shake his questions as to just *who* that first man had been.  

At nineteen he sat for weeks, alone in his bedroom poring over a large stack of books and compiling hundreds of seemingly deranged and disjointed notes. When he emerged, he ventured into the wilderness nearby and found that very same man in a small cabin, deep in the obscure forest. At the very least he knew that man was indeed real and not just some figment within his imagination. This revelation had troubled him deeply. When the man asked how he'd come to find his cabin, Brian had said that he'd simply "Followed The Stars."  

By twenty-one he'd grown quite comfortable with the apparently hidden realm of knowledge he'd somehow stumbled upon which seemed so contradictory toward what most considered "reality." Indeed, there had often been times throughout Brian's life when random things—seemingly unrelated events—would occur in accordance with his thoughts, as though everything around him knew him personally. For the most part, by now he'd learned it best simply to ignore such sensations. Never mind the fact he could no longer explain away the existence of the man in the cabin—he'd realized rather painfully, however, that far too often following after such strange thoughts always left him at a dead-end. A barren space—wherein things were neither resolved, nor truly broken.  

He'd also learned it was best to keep quiet about such things after one evening's chat with his sisters—one older, the other younger. They'd immediately pointed him in the direction of a psychiatric ward, citing various articles on the dangers of *schizophrenia*. He actually had taken their advice at one point. A time when his mind seemed to be unraveling by the seams, but when he'd told the doctor he feared he was going insane, the man had laughed at him, then reassuringly said, "No my boy, you see, crazy people don't *know* they're crazy," and suggested he simply try to get more sleep.  

Brian tried to take everything in stride, and despite possessing a very powerful "sixth sense" or intuitive notion about *most* things, he now rarely let himself become too enmeshed with the narrative's inherent drama. He routinely ignored the underlying strangeness of the world around him, blending it all in rather nicely with the rest of the odd characters he encountered—people who seemed to know him in some fashion or another—which is why his reaction on *this* day was quite out of the ordinary indeed.  

He began running after the jogging experiment failed to bring him *any* closer to the man. He was now convinced there was *indeed* some sort of gulf fixed between them, as if—dare he say—by magic, as it were. The pursuit seemed to go on for at least a good half-hour. He'd tried everything he could to close or even extend the gap, but it was all in vain. Once Brian even stopped entirely but it too was to no avail. The man just kept on going, yet never seemed to get any further away. Brian curiously found himself suddenly actually *wanting* to talk to the stranger after all. *This* was most *definitely* out of character for the young man.  

After leading Brian to a rather poor-looking side of town, the man turned swiftly down an alleyway, in between two apartment buildings. As Brian rounded the corner, he saw the man open a door, then slam it closed behind him. His eyes widened. He made his way past the dangling clothesline and bits of scattered trash that littered the alleyway, and then came across yet *another* bewitching and auspicious symbol. He stared, transfixed by the numbers which glared back at him.  

Throughout Brian's odd life, a tapestry of patterned and coded number schemes inexplicably seemed to not only appear but also somehow held a specific meaning he had no rational means of explaining. The timing of their appearances always emphasized this in dramatic and impossible ways. This phenomenon did nothing to change what he felt, nor dull the shocking intensity this particular configuration inflicted, however. The number in question was one that had haunted him ever since his mother died. He was only sixteen at the time and she had just turned forty. You see, the apartment was labeled six-hundred fourteen, and to Brian, any combination of the three numbers—one, six, and four—had always said to him the same thing. *Death!* plain and simple.  

He thought for a brief moment of leaving, returning to his car and going home, but luckily for us all, his ego urged him onward. Without any further hesitation or outward show of fear, he confidently turned the handle, only to find the door dead-bolted. This baffled him profusely, and after giving the knob another futile turn, he decided instead to knock, unsure suddenly if the man he'd followed even existed.  

On his third knock, and almost in stride with his hand, the door swung open and a different voice met him.  

"Ah, Mr. Johnson, it's so nice to *finally* meet you. Please, won't you come inside?"  

This was not the voice of the man he'd followed. No, this was not the same man at all.  

"We've been waiting for you," a familiar and snakelike voice concluded.  

As Brian peered inside he saw the man with the green eyes. He sat in a large velvet recliner and was smiling wickedly back at him.  

"Please, come in," said the man at the door once more. He was short and chubby—wearing the oddest of clothes. A full tuxedo.  

Brian was now unsettled to say the least. In fact, he'd secretly hoped it was indeed all just a hallucination. The man he had followed was not real at all, and the whole thing had been a farce—yet there he sat, just inside, smiling sinisterly. He again began to doubt his choice to ever abandon his breakfast. He stood there hesitantly, attempting to assess the safety of the situation from behind his dark sunglasses. He naturally felt as though he were being deceived in some fashion.  

The man in the tux, though, was insistent and began gesturing with his hand. "Please Mr. Johnson, *do* come inside, we have *much* to discuss."  

"How do you know my name?" asked Brian sternly, once more on the defensive.  

The man had such an inviting face, round and comforting, upon it wearing a warm smile, but when Brian said this, the smile just... disappeared.  

"Yes! How *does* he know your *name* Mr. McCord?" hissed the green-eyed man from his chair. "Come in, please, so Elvis can shut the door!"  

Brian laughed to himself, though his face did not show it. The man at the door indeed looked like a shorter, fatter rendition of Elvis. However, the irony of the situation didn't lower Brian's guard in the slightest. If anything, it caused him to be all the more cautious. These men knew both his real name and alias—he'd put quite a bit of energy and money into keeping the correlation from becoming public knowledge. They seemed to be mocking him about it. He wasn't yet fully persuaded, and he had a sense their intentions weren't as pure as they seemed to be trying so desperately to suggest—though they also seemed not to care much about hiding their subterfuge either.  

Feeling suddenly as if he could reassert control, Brian decided to test their resolve.  

"No... You tell me *now* how both of you seemingly know me... or I'm going home."  

Much to his surprise, they responded rather oppositely than he expected. You see, Brian thought they needed *him* for something, and the idea had not yet crossed his mind that it in fact was the other way around.  

The two men looked at each other, shrugged in unison, and said, "Have it your way."  

Elvis slammed the door shut.  

Brian heard the deadbolt slide back into place, and after a few brief seconds of confusion, knocked once more. *Again, a most unnatural choice of actions on his part.* Normally, despite the strangeness, Brian actually would have just left, purposefully forgetting about the whole affair, but *something* seemed to refuse to allow him this privilege.  

When the door opened again, Brian wasted no time before confidently striding inside. He said nothing, and when the door shut behind him the room grew dark, forcing him to remove his sunglasses. The only light in fact was a small lamp sitting on an end table, situated in the far corner between the recliner and sofa. The room was neat—*too neat* it seemed, especially for this side of town. The books upon the shelves were all in alphabetical order and the white high-pile carpet was pristine. The light from the lamp bounced off the walls, giving the room an eerie yellowish glow. The window had been boarded up, and over it hung a portrait of a dashingly beautiful woman dressed in a thin white gown, her long, silvery-blonde hair brushed neatly over the front of her shoulders. The parquet inlay near the door seemed polished as it gleamed in the dim light.  

"Oh please, do have a seat Mr. McCord, there are some things we would like to go over with you," said the stranger with the green eyes, gesturing him towards the couch. "And don't worry, we'll tell you *everything* you want to know, but first," the man pointed once more, "You must sit."  

"Would you like some tea Mr. Johnson?" the fat man asked.  

Brian made his way across the room. "No *Elvis*, but thank you," he replied, a hint of sarcasm in his tone. He glanced up at the portrait over the window as he passed, feeling a shiver run first down, and then back up his spine as he did. It seemed as though her gaze had pierced through to his very soul, and she seemed... vaguely familiar.  

It was clear though the apartment was situated in the very worst part of town, no poor man lived here. Odder still, as Brian sat down and sank into his seat, he also got the sense that though everything *looked* new, it had all been there for a very, very long time.  

"Well, now that you're seated, and don't want any tea, let's get down to business, shall we?" asked the stranger in the recliner, his green eyes flickering in the dim light, smiling evilly.  

"Yes..." Brian said hesitantly. "Let's..."  

At this point the man began spewing forth a torrent of words, as though he'd been waiting for a *very long time* for this exact moment, and though Brian had the same exaggerated look of boredom upon his face, he was *quite* interested, but only because he was suddenly overcome with a very intense feeling of déjà vu as he stared at the man.  

"First things first, that is not *The Elvis*. I know he resembles him in an odd sort of way, but just get that out of your head. You're not crazy. Remember that." Which was an odd thing to hear, because at that very moment, Brian had begun to very seriously question that very thing.  

"My name," the stranger continued, placing his hands dramatically upon his chest, "is Ricco, and yours is Brian McCord, Tracy Johnson, or whatever other ridiculous thing it is that you want to be called."  

The man's somewhat ecstatic manner set Brian on edge once more. He interrupted his speech.  

"Well, that about handles introductions. Do you want to tell me why it is that I followed you?" Brian asked.  

"Ah yes, that is an excellent question." He leaned forward, an eyebrow cocked as he smiled back wickedly at Brian. "Why *did* you follow me?"  

Brian was thrown off by this and for the first time it showed upon his face as his eyes went wide.  

The man eyed him curiously for a moment, and then suddenly, with great excitement he said, "I think it's time we finally told him everything, Elvis!"  

"Yes... Yes it is Fenningway. As I always say, strike while the iron is hot!"  

Brian caught himself. Where had he heard that name before? Ricco Fenningway... he thought to himself.  

"Fenningway...? Did you... Did you just say, Fenningway, as in A.R. Fenningway?"  

Ricco stared back at Brian, his smile gone, his face assuming a look of dire intensity.  

"Yes Mr. McCord," he began, even-toned, staring down his nose. "A. R. Fenningway. Alfred Ricco Fenningway to be exact, your sixth-grade math and science teacher. Do you remember me now?"  

Brian noticed that as the man leaned back into his recliner, he cut his eyes toward the portrait over the window, ever so subtly.  

"But... you don't look anything like the man I once knew."  

"And you don't *look* anything like an old student of mine, but you are and I am." He said in a tone that seemed disgusted with Brian's ignorance.  

Poor Brian was now more confused than he thought he had been already, but he didn't let on to it just yet. He casually returned to his bored demeanor. "Whatever do you mean, Mr. Fenningway?" he asked, sounding perturbed.  

Elvis met Brian's wandering gaze as he looked away from the chair, smiling brightly and suddenly at him when he did, seemingly overjoyed about something.  

"Don't you call me *mister* anything!" screamed Fenningway, pointing a finger warningly, directly into Brian's face. Again, Brian didn't flinch, although most people would have. He hadn't seen Fenningway move. One moment the man was calmly sitting back in the recliner, and the next he was mere inches from Brian's face, seething with anger.  

"Sit *down*, Ricco! Sit down!" barked Elvis, a new tone darkening his previously cheerful voice.  

"I have to apologize for my friend, you see, he hates for anyone to show him any sort of respect or common courtesy, such as saying *mister* or referring to him as *sir*."  

"So do I..." said Brian curiously.  

"Well then, there we have it!" exclaimed Fenningway, jumping up once more, finger raised high into the air this time.  

"I said sit down!" Elvis roared.  

Suddenly the warm and inviting face seemed to be *much* more of a disguise than a reality.  

"Oh... yes," said Fenningway sheepishly. "Sorry about that."  

Elvis' demeanor shifted back to that of friendly, and Brian marveled secretly at the suddenness of it.  

"My name," Elvis continued, "is Elvis Harker. I haven't ever met you before, but Fenningway here has. There's a lot to explain, and not much time to do so, so I must insist that you keep your questions to a minimum, as I'm sure we'll get to them sooner or later, and those that we don't will simply answer themselves anyway in their own time." The man smiled brightly at him once more.  

Brian nodded. He was now sure that *these two men* were insane, not himself.
